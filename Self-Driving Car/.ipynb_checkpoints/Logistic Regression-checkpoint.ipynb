```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
n_pts = 100        # number of points for the dataset
np.random.seed(0)  # Ensures random points are generated ever code launch
top_region = np.array([np.random.normal(10, 2, n_pts), np.random.normal(12,2,n_pts)]).T
bot_region = np.array([np.random.normal(5, 2, n_pts), np.random.normal(6,2,n_pts)]).T

_, ax = plt.subplots(figsize=(4,4))
ax.scatter(top_region[:, 0], top_region[:, 1], color = 'r')
ax.scatter(bot_region[:, 0], bot_region[:, 1], color='b')
plt.show()
```
